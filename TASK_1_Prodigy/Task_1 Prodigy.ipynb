```python
#Importing the necessary Libraries and Modules
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#Assigning the file path to a variable
file_path = "E:\COMPUTER SCIENCE\PRODIGY_INTERNSHIP\TASK_1_Prodigy\CSV_Task_1_Prodigy\data_task_1_prodigy.csv"
#Using Pandas to read the CSV file
df = pd.read_csv(file_path)
```

```python
#Select a column name to plot here we have choosen '1998'
choosen_year = '1998'
# x axis is taken as population & y axis is Country name.
data_for_year = df[['Country Name', choosen_year]]
#Ascending (denotes the order preference can be True/False)
data_for_year = data_for_year.sort_values(by = choosen_year, 
                                          ascending = True)
#Plotting the top 25 alone 
data_for_year_top25 = data_for_year.head(25)
plt.figure(figsize = (12,10))
sns.barplot(y = 'Country Name', 
            x = choosen_year, 
           data = data_for_year_top25, 
            orient = 'h')
plt.title('TOP 25(1998)')
plt.xlabel('Population')
plt.ylabel('Country Name')
plt.show()
```

```python
"""
We could see Aruba has the highest population &
Tuvalu has the least as per data for the year 1998"""
```

```python
#Plotting for 
choosen_year = '2019'
# x axis is taken as population & y axis is Country name.
data_for_year = df[['Country Name', choosen_year]]
#Ascending (denotes the order preference can be True/False)
data_for_year = data_for_year.sort_values(by = choosen_year, 
                                          ascending = False)
#Plotting the top 10 alone 
data_for_year_top10 = data_for_year.head(10)
plt.figure(figsize = (12,10))
sns.barplot(y = 'Country Name', 
            x = choosen_year, 
           data = data_for_year_top10, 
            orient = 'h')
plt.title('TOP 10(2019)')
plt.xlabel('Population')
plt.ylabel('Country Name')
plt.show()
```

```python
"""
We could see World has the highest population &
Late-demographic divident has the least as per data for the year 2019"""
```

```python
#Comparing two unique years using Seaborn.
year_1960 = '1960'
year_2023 = '2023'
data_selected_years = df[['Country Name',year_1960,year_2023]]
plt.figure(figsize = (10,6))
sns.regplot(x = year_1960, y = year_2023, data = data_selected_years, 
           line_kws = {'color': 'red'})
plt.title(f'Population in {year_1960} VS {year_2023}')
plt.xlabel("1960")
plt.ylabel('2023')
plt.show()
```
